Time-series decompositon

It allows us to decompose our time series into three distinct components

->Trend

->Seasonality

->Noise

```python
import statsmodels.api as sm
decompo = sm.tsa.seasonal_decompose(res1, model = 'additive')
fig = decompo.plot()
plt.show()
```

We can say that the sales of the company...,
1) The trend is increasing so the sales of the company is growing periodically

2) The seasonal index shows the sales every second half of the year the sales is incresed seasonally

# Forecasting with ARIMA model

```python
p = d = q = range(0, 2)
model = sm.tsa.statespace.SARIMAX(res1, order = (1, 1, 1),
                                  seasonal_order = (1, 1, 0, 12)
                                 )
result = model.fit()
print(result.summary().tables[1])
```

```python
prediction = result.get_prediction(start = pd.to_datetime('2017-01-01'), dynamic = False)
prediction_ci = prediction.conf_int()
prediction_ci
```

```python
ax = res1['2014':].plot(label = 'observed')
prediction.predicted_mean.plot(ax = ax, label = 'One-step ahead Forecast', alpha = 0.7, figsize = (14, 7))
ax.fill_between(prediction_ci.index, prediction_ci.iloc[:, 0], prediction_ci.iloc[:, 1], color = 'k', alpha = 0.2)
ax.set_xlabel("Date")
ax.set_ylabel('Furniture Sales')
plt.legend()
plt.show()
```

```python
y_hat = prediction.predicted_mean
y_truth = res1['2017-01-01':]

mse = ((y_hat - y_truth) ** 2).mean()
rmse = np.sqrt(mse)
print('The Mean Squared Error of our forecasts is {}'.format(round(mse, 2)))
print('The Root Mean Squared Error of our forecasts is {}'.format(round(rmse, 2)))
```

```python

# forcasting for out of sample data
pred_uc = result.get_forecast(steps = 100)
pred_ci = pred_uc.conf_int()

ax = res1.plot(label = 'observed', figsize = (14, 7))
pred_uc.predicted_mean.plot(ax = ax, label = 'forecast')
ax.fill_between(pred_ci.index, pred_ci.iloc[:, 0], pred_ci.iloc[:, 1], color = 'k', alpha = 0.25)
ax.set_xlabel('Date')
ax.set_ylabel('Furniture Sales')

plt.legend()
plt.show()
```

Thus we predicted the averae sales for the future years
